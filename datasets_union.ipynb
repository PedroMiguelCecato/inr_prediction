## Lendo todos os pacientes registrados nas planilhas

```python
import pandas as pd
import glob
import os
from models.data_manipulator import INRDataManipulation
from sklearn.model_selection import train_test_split
```

```python
caminho = r"G:\.shortcut-targets-by-id\1pVv4_awj3MCrGLaPaYx5OLlzmRd51R8S\GemeosDigitais_Saude\Planilhas HNE\*.xlsx" 

arquivos = glob.glob(caminho)

df_weekly_full = []
df_weekly_train = []
df_weekly_test = []

df_final_full = []
df_final_train = []
df_final_test = []

for arquivo in arquivos:
    try:        
        paciente = INRDataManipulation(path=arquivo)
        data_weekly = paciente.get_data_weekly()
        data_final = paciente.get_data_final()

        cutoff_date = data_weekly["test_date"].quantile(0.8)
        # split_index_w = int(len(data_weekly) * 0.8)
        # split_index_f = int(len(data_final) * 0.8)

        df_weekly_full.append(data_weekly.copy())
        df_weekly_train.append(data_weekly[data_weekly["test_date"] < cutoff_date])
        df_weekly_test.append(data_weekly[data_weekly["test_date"] >= cutoff_date])

        df_final_full.append(data_final.copy())
        df_final_train.append(data_final[data_final["test_date"] < cutoff_date])
        df_final_test.append(data_final[data_final["test_date"] >= cutoff_date])

    except Exception as e:
        print(f"Erro ao processar {arquivo}: {e}")
```

```python
df_weekly_full = (
    pd.concat(df_weekly_full, ignore_index=True)
      .sort_values(by=['nome_paciente', 'test_date'])
      .reset_index(drop=True)
)

df_weekly_train = (
    pd.concat(df_weekly_train, ignore_index=True)
      .sort_values(by=['nome_paciente', 'test_date'])
      .reset_index(drop=True)
)

df_weekly_test = (
    pd.concat(df_weekly_test, ignore_index=True)
      .sort_values(by=['nome_paciente', 'test_date'])
      .reset_index(drop=True)
)

df_final_full = (
    pd.concat(df_final_full, ignore_index=True)
      .sort_values(by=['nome_paciente', 'test_date'])
      .reset_index(drop=True)
)

df_final_train = (
    pd.concat(df_final_train, ignore_index=True)
      .sort_values(by=['nome_paciente', 'test_date'])
      .reset_index(drop=True)
)

df_final_test = (
    pd.concat(df_final_test, ignore_index=True)
      .sort_values(by=['nome_paciente', 'test_date'])
      .reset_index(drop=True)
)
```

## Removendo valores nulos e formulando novos datasets

```python
df = df_weekly_full.copy()

# Agrupamento por paciente
resumo_missing_weekly = (
    df.groupby("nome_paciente")
      .agg(
          total_registros=("dose_semanal", "size"),
          n_nan=("dose_semanal", lambda x: x.isna().sum()),
          n_validos=("dose_semanal", lambda x: x.notna().sum())
      )
)

# Calcular percentual
resumo_missing_weekly["percent_nan"] = (
    resumo_missing_weekly["n_nan"] / resumo_missing_weekly["total_registros"]
) * 100

# Ordenar do pior para o melhor
resumo_missing_weekly = resumo_missing_weekly.sort_values(
    by="percent_nan",
    ascending=False
)

display(resumo_missing_weekly.head(50))
```

```python
print("Total de pacientes registrados:", pd.unique(df_weekly_full["nome_paciente"]).shape[0])

pacientes_remover_weekly = resumo_missing_weekly[
    resumo_missing_weekly["percent_nan"] > 30
].index.tolist()

print("Pacientes a remover (weekly):", len(pacientes_remover_weekly))

df_weekly_full_filtred = df_weekly_full[~df_weekly_full["nome_paciente"].isin(pacientes_remover_weekly)]
df_weekly_train_filtred = df_weekly_train[~df_weekly_train["nome_paciente"].isin(pacientes_remover_weekly)]
df_weekly_test_filtred = df_weekly_test[~df_weekly_test["nome_paciente"].isin(pacientes_remover_weekly)]
```

```python
print("Tamanho df_weekly_full_filtred:", len(df_weekly_full_filtred))
print("Tamanho df_weekly_train_filtred:", len(df_weekly_train_filtred))
print("Tamanho df_weekly_test_filtred:", len(df_weekly_test_filtred), "\nValores nulos:")
print(df_weekly_full_filtred.isna().sum())
```

```python
print(df_weekly_test_filtred.isna().sum())
```

```python
df_weekly_full_filtred = df_weekly_full_filtred.dropna()
df_weekly_train_filtred = df_weekly_train_filtred.dropna()
df_weekly_test_filtred = df_weekly_test_filtred.dropna()
```

```python
print("Tamanho df_weekly_full_filtred:", len(df_weekly_full_filtred))
print("Tamanho df_weekly_train_filtred:", len(df_weekly_train_filtred))
print("Tamanho df_weekly_test_filtred:", len(df_weekly_test_filtred), "\nValores nulos:")
print(df_weekly_full_filtred.isna().sum())
```
